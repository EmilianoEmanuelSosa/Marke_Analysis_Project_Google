```python
import pandas as pd
```

```python
csv_file_path_restaurants = '/home/mkm/programin/Marke_Analysis_Project_Google/Data/Data_process/restaurants.csv'
csv_file_path_users = '/home/mkm/programin/Marke_Analysis_Project_Google/Data/Data_process/users.csv'
csv_file_path_reviews = '/home/mkm/programin/Marke_Analysis_Project_Google/Data/Data_process/reviews.csv'
```

```python
df_restaurant = pd.read_csv(csv_file_path_restaurants)
df_users = pd.read_csv(csv_file_path_users)
df_reviews = pd.read_csv(csv_file_path_reviews)
```

```python
df_reviews.drop_duplicates(subset=['user_id'], inplace=True)
```

```python
df_users.drop_duplicates(subset=['user_id'], inplace=True)
```

```python
df_restaurant.drop_duplicates(subset=['restaurant_id'],inplace=True)
```

```python
df_reviews
```

```python
# Unir df_reviews con df_users en base a la columna user_id
merged_df = df_reviews.merge(df_users, on='user_id', how='inner')

# Unir el DataFrame combinado anterior con df_restaurant en base a la columna restaurant_id
merged_df = merged_df.merge(df_restaurant, on='restaurant_id', how='inner')
```

```python
merged_df.drop(columns=['name_y','name_x'],axis=1,inplace=True)
```

```python
merged_df.drop(columns=['address'],axis=1,inplace=True) 
```
